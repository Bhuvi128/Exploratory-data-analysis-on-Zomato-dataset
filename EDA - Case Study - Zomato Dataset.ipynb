## Problem Statement

The Zomato exploratory data analysis aims to assist food enthusiasts in discovering the best restaurants and value-for-money options within their locality. It facilitates locating desired cuisines available nearby. With a dataset encompassing restaurant details like establishment type, address, city, locality, and cuisine offerings, alongside attributes such as average cost for two, price range, and aggregate ratings, the objective is to uncover insights into restaurant preferences and dining trends. Through analyzing factors like average cost, ratings, and customer engagement metrics such as votes and photo counts, this study seeks to provide actionable recommendations for diners seeking satisfying dining experiences.

## Data Definition

**res_id**: The code given to a restaurant (Categorical) 

**name**: Name of the restaurant (Categorical)

**establishment**: Represents the type of establishment (Categorical)

**url**: The website of the restaurant (Categorical)

**address**: The address of the restaurant (Categorical)

**city**: City in which the restaurant located (Categorical)

**city_id**: The code given to a city (Categorical)

**locality**: Locality of the restaurant (Categorical)

**latitude**: Latitude of the restaurant (Categorical)

**longitude**: Longitude of the restaurant (Categorical)

**zipcode**: Zipcode of the city in which the restaurant located (Categorical)

**country_id**: Country code in which the restaurant located (Categorical)

**locality_verbose**: Locality along with the city in which the restaurant located (Categorical)

**cuisines**: The cuisines a restaurant serves (Categorical)

**timings**: The working hours of a restaurant (Categorical)

**average_cost_for_two**: The average amount expected for 2 people (Numerical)

**price_range**: The categories for average cost (Categories - 1,2,3,4) (Categorical)

**currency**: The currency in which a customer pays (Categorical)

**highlights**: The facilities of the restaurant (Categorical)

**aggregate_rating**: The overall rating a restaurant has got (Numerical) 

**rating_text**: Categorized ratings (Categorical)

**votes**: Number of votes received by the restaurant from customers (Numerical)

**photo_count**: The number of photos of a restaurant (Numerical)

**opentable_support**: Restaurant reservation from Opentable (Categorical)

**delivery**: The restaurant deliver an order or not (Categorical)

**takeaway**: The restaurant allows  a 'takeaway' of an order or not (Categorical)

## 8. Detect outliers

**1. Plot the boxplot for numeric data**

```python
numerical_df = df_cleaned.select_dtypes(include=['float64', 'int64'])


plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_df.columns, 1):
    plt.subplot(len(numerical_df.columns), 1, i)
    sns.boxplot(x=numerical_df[column])
    plt.title(f'Boxplot of {column}')


plt.show()
```

**2. Note the variables for which outliers are present**

```python
numerical_df = df_cleaned.select_dtypes(include=['float64', 'int64'])


def detect_outliers(df):
    outlier_info = {}
    
    for column in df.columns:
        if df_cleaned[column].dtype in ['float64', 'int64']:
            Q1 = df_cleaned[column].quantile(0.25)
            Q3 = df_cleaned[column].quantile(0.75)
            IQR = Q3 - Q1
            

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            
            outliers_count = df_cleaned[(df[column] < lower_bound) | (df_cleaned[column] > upper_bound)][column].count()
            
            if outliers_count > 0:
                outlier_info[column] = outliers_count

    return outlier_info


outlier_variables = detect_outliers(numerical_df)


print(outlier_variables)
```

**3. Plot the boxplot to recheck for outliers**

```python
numerical_df = df_cleaned.select_dtypes(include=['float64', 'int64'])


plt.figure(figsize=(15, 10))


for i, column in enumerate(numerical_df.columns, 1):
    plt.subplot(len(numerical_df.columns), 1, i)
    sns.boxplot(x=numerical_df[column])
    plt.title(f'Boxplot of {column}')


plt.show()
```

## 10. Bivariate and Multi Variate Analysis

```python
# Relationship between average_cost_for_two and aggregate_rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='average_cost_for_two', y='aggregate_rating')
plt.title('Scatter Plot: Average Cost for Two vs Rating')
plt.xlabel('Average Cost for Two')
plt.ylabel('Rating')
plt.show()

# Calculate the correlation coefficient
correlation = df_cleaned['average_cost_for_two'].corr(df_cleaned['aggregate_rating'])
print(f"Correlation between average_cost_for_two and aggregate_rating: {correlation}")
```

```python
# Distribution of Average cost for two

plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
sns.histplot(df_cleaned['average_cost_for_two'], kde=True, bins=30, color='blue')
plt.title('Distribution of Average Cost for Two')
plt.xlabel('Average Cost for Two')

plt.show()
```

```python
# Distribution of Price Range

plt.figure(figsize=(16,4))

plt.subplot(1, 3, 2)
sns.countplot(x='price_range', data=df_cleaned, palette='Set2')
plt.title('Distribution of Price Range')
plt.xlabel('Price Range')

plt.show()
```

```python
# Distribution of Aggregate Ratings

plt.figure(figsize=(16, 4))


plt.subplot(1, 3, 3)
sns.histplot(df_cleaned['aggregate_rating'], kde=True, bins=30, color='green')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')

plt.show()
```

```python
# Top rated restaurants

top_rated_restaurants = df_cleaned.sort_values(by=['aggregate_rating', 'votes'], ascending=False)

# Display the relevant details
top_rated_restaurants.head()
```

```python
# Cuisine Preferences

# Split the 'cuisines' column into individual cuisines (as some restaurants offer multiple cuisines)
df_cleaned['cuisines'] = df_cleaned['cuisines'].fillna('Unknown')
df_cleaned['cuisines_split'] = df_cleaned['cuisines'].str.split(',')

# Create a new DataFrame where each row corresponds to a single cuisine for a restaurant
cuisine_explode = df_cleaned.explode('cuisines_split')

# Clean the 'cuisines_split' column by stripping whitespace
cuisine_explode['cuisines_split'] = cuisine_explode['cuisines_split'].str.strip()

# Calculate cuisine frequency
cuisine_frequency = cuisine_explode['cuisines_split'].value_counts().head(10)

# Calculate average rating by cuisine
cuisine_avg_rating = cuisine_explode.groupby('cuisines_split')['aggregate_rating'].mean().sort_values(ascending=False).head(10)
```

```python
# Plotting popular cuisines by frequency
plt.figure(figsize=(10, 6))
sns.barplot(x=cuisine_frequency.values, y=cuisine_frequency.index, palette='Blues_r')
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Frequency')
plt.ylabel('Cuisine')
plt.show()
```

```python
# Plotting top-rated cuisines by average rating
plt.figure(figsize=(10, 6))
sns.barplot(x=cuisine_avg_rating.values, y=cuisine_avg_rating.index, palette='Greens_r')
plt.title('Top 10 Highest Rated Cuisines')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')
plt.show()
```

```python
# Scatter plot: Votes vs Aggregate Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='votes', y='aggregate_rating', data=df_cleaned, hue='price_range', palette='coolwarm')
plt.title('Votes vs Aggregate Rating')
plt.xlabel('Votes')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
# Scatter plot: Photo Count vs Aggregate Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='photo_count', y='aggregate_rating', data=df_cleaned, hue='price_range', palette='coolwarm')
plt.title('Photo Count vs Aggregate Rating')
plt.xlabel('Photo Count')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
# Correlation analysis for votes, photo_count, and aggregate_rating
engagement_correlation = df_cleaned[['votes', 'photo_count', 'aggregate_rating']].corr()

# Display the correlation matrix
engagement_correlation
```

```python
# Heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(engagement_correlation, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title('Correlation between Customer Engagement Metrics and Aggregate Rating')
plt.show()
```
